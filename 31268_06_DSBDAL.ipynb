```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import set_theme
```

```python
df =  pd.read_csv('datasets/Iris.csv')
```

```python
df.head(10)
```

```python
df.dtypes
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 10))

fig.suptitle('Feature Distribution by Class')

sns.histplot(ax=axes[0, 0], data=df, x='SepalLengthCm', hue='Species')
sns.histplot(ax=axes[0, 1], data=df, x='SepalWidthCm', hue='Species')
sns.histplot(ax=axes[1, 0], data=df, x='PetalLengthCm', hue='Species')
sns.histplot(ax=axes[1, 1], data=df, x='PetalWidthCm', hue='Species')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 10))

fig.suptitle('Feature Distribution by Class')

sns.boxplot(ax=axes[0, 0], data=df, x='SepalLengthCm', y='Species', hue='Species')
sns.boxplot(ax=axes[0, 1], data=df, x='SepalWidthCm', y='Species', hue='Species')
sns.boxplot(ax=axes[1, 0], data=df, x='PetalLengthCm', y='Species', hue='Species')
sns.boxplot(ax=axes[1, 1], data=df, x='PetalWidthCm', y='Species', hue='Species')
```

### Scatter Plot

#### 1) Sepal Length vs Sepal Width

```python
sns.scatterplot(data = df, x = 'Sepal_Length', y = 'Sepal_Width', hue = 'Class')
```

#### 2) Petal Length vs Petal Width

```python
sns.scatterplot(data = df, x = 'Petal_Length', y = 'Petal_Width', hue = 'Class')
```

### Conclusions :

1) For Iris-setosa sepal-length, petal-length, petal-width is smaller and for Iris-virginica sepal-length, petal-length, petal-width is greater.
2) Sepal-width of Iris-setosa is greater than other two species.
3) From scatter plot and boxplot, it is clear that Iris-setosa has smaller length ratio of petal-length and petal-width.
4) From scatter plot and boxplot, it is seen that Iris-virginica has greater length ratio of petal-length and petal-width.
5) Iris-versicolor shows average features.
