```python
import pandas as pd
```

```python
# Load the dataset
dataframe = pd.read_csv('nba.csv')
dataframe.isnull().sum()
```

```python
dataframe.dropna(inplace=True)
dataframe.info()
```

```python
dataframe.groupby(['Team'])['Weight'].mean()
```

```python
dataframe['Team'].value_counts()
```

```python
dataframe.groupby(['Age', 'Salary']).describe()
```

```python

bins = [0, 20, 30, 40, 50, 100]
labels = ['>20','20-30', '30-40', '40-50', '<50']
dataframe['Age Group'] = pd.cut(x=dataframe['Age'], bins=bins, labels=labels)
dataframe
```

```python
dataframe.groupby(['Age Group']).describe()
```
