```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
# countplot
# countplot is used to show the counts of observations in each categorical bin using bars.

sns.countplot(x = 'Pclass', data=df)
plt.show()

sns.countplot(x='Sex', data=df)
plt.show()

sns.countplot(x='Survived', data=df, hue='Sex')
plt.show()
```

```python
# Line plot
# Line plot is used to plot data points in a line. It is used to show the relationship between two variables.

sns.lineplot(x='Age', y='Fare', data=df, hue="Sex")
# hue is used to differentiate the line based on the column specified. In the above example the line is differentiated based on sex
```

```python
# Bar plot
# Bar plot is used to show the relationship between a numerical and a categorical variable. 
# It is used to show the average value of the numerical variable for each category.

sns.barplot(x='Pclass', y='Fare', data=df)
plt.show()

sns.barplot(x='Pclass', y='Fare', data=df, hue='Sex')
plt.show()
```

```python
# Histogram
# Histogram is used to show the distribution of a numerical variable. It is used to show the frequency of the variable in each bin.

sns.histplot(x='Age', data=df)
plt.show()

sns.histplot(x='Age', data=df, bins=8) # bins is used to specify the number of bins in the histogram
plt.show()

sns.histplot(x='Age', data=df, kde=True) # kde is used to show the kernel density estimation
plt.show()
```

```python
# Scatter plot
# Scatter plot is used to show the relationship between two numerical variables. It is used to show the correlation between the two variables. 

sns.scatterplot(x='Age', y='Fare', data=df)
plt.show()

sns.scatterplot(x='Age', y='Fare', data=df, hue='Sex')
plt.show()
```

```python
# Box plot
# Box plot is used to show the distribution of a numerical variable. It is used to show the median, quartiles, and outliers of the variable. 

sns.boxplot(y='Fare', data = df)
plt.show()

sns.boxplot(x='Pclass', y='Fare', data = df)
plt.show()

sns.boxplot(data=df, x='Survived', y='Age', hue='Sex')
plt.show()
```
