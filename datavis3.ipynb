```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
# 1. List down the features and their types (e.g., numeric, nominal) available in the dataset.

print(df.dtypes)
print(df.describe())
```

```python
# 2. Create a histogram for each feature in the dataset to illustrate the feature distributions.

df.hist()
plt.show()

sns.histplot(x='SepalLengthCm', data=df, kde=True)
plt.show()

sns.histplot(x='SepalWidthCm', data=df, kde=True)
plt.show()

sns.histplot(x='PetalLengthCm', data=df, kde=True)
plt.show()

sns.histplot(x='PetalWidthCm', data=df, kde=True)
plt.show()
```

```python
# 3. Create a box plot for each feature in the dataset. Compare distributions and identify outliers.

sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.show()

sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.show()

sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.show()

sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.show()
```
