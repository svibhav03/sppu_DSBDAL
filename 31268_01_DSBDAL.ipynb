```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/melb_data.csv")
data
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data['Car'].fillna(data['Car'].mean(), inplace = True)
data['BuildingArea'].fillna(data['BuildingArea'].mean(), inplace = True)
data['YearBuilt'].fillna(data['YearBuilt'].mean(), inplace = True)
data['CouncilArea'].fillna(data['CouncilArea'].mode()[0], inplace = True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.dtypes
```

```python
data['Date'] = data["Date"].astype("datetime64")
data['Postcode'] = data['Postcode'].astype('int64')
data['Bedroom2'] = data['Bedroom2'].astype('int64')
data['Bathroom'] = data['Bathroom'].astype('int64')
data['Car'] = data['Car'].astype('int64')
data['YearBuilt'] = data['YearBuilt'].astype('int64')
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['CouncilArea'].unique()
```

```python
labelencoder_CouncilArea = LabelEncoder()
data['CouncilArea'] = labelencoder_CouncilArea.fit_transform(data['CouncilArea'])
data.head()
```

```python
data['CouncilArea'].unique()
```

```python
labelencoder_SellerG = LabelEncoder()
data['SellerG'] = labelencoder_SellerG.fit_transform(data['SellerG'])

labelencoder_Method = LabelEncoder()
data['Method'] = labelencoder_Method.fit_transform(data['Method'])
data.head()
```

```python
data = pd.get_dummies(data, columns=['Method'])
data.head()
```
